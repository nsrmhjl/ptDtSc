###<div dir =rtl><h3><font color = blue>تاپل لانه سازی شده</font></h3></div>


<div dir =rtl><p>یک تاپل می تواند  حاوی یک تاپل دیگر و نیز انواع دیگری از داده های پیچیده را ذخیره کند. به این روند اصطلاحا لانه سازی می گویند. تاپل زیر با چند جزء را در نظر بگیرید:</p></div>



```python
# Create a nest tuple

NestedT =(1, 2, ("pop", "rock") ,(3,4),("disco",(1,2)))
```

<div dir =rtl><p>هر جزء در تاپل را می توان با یک اندکس به دست آورد. این شامل تاپل های دیگر هم می شود. همانطور که در شکل نمایش داده شده است:</p></div>


TuplesNestOne.png

```python
# Print element on each index

print("Element 0 of Tuple: ", NestedT[0])
print("Element 1 of Tuple: ", NestedT[1])
print("Element 2 of Tuple: ", NestedT[2])
print("Element 3 of Tuple: ", NestedT[3])
print("Element 4 of Tuple: ", NestedT[4])
```

<div dir =rtl><p>ما می توانیم از جزء دوم برای دسترسی به تاپل های لانه سازی شده استفاده کنیم . همانگونه که در شکل نشان داده شده است:</p></div>


<img src="https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/PY0101EN/Chapter%202/Images/TuplesNestTwo.png" width="750" align="center">

 We can access the nested tuples :

```python
# Print element on each index, including nest indexes

print("Element 2, 0 of Tuple: ",   NestedT[2][0])
print("Element 2, 1 of Tuple: ",   NestedT[2][1])
print("Element 3, 0 of Tuple: ",   NestedT[3][0])
print("Element 3, 1 of Tuple: ",   NestedT[3][1])
print("Element 4, 0 of Tuple: ",   NestedT[4][0])
print("Element 4, 1 of Tuple: ",   NestedT[4][1][1])
```

<div dir =rtl><p>ما می توانیم به رشته حروف موجود در تاپل دوم لانه سازی شده دسترسی پیدا کنیم:</p></div>


```python
# Print the first element in the second nested tuples

NestedT[2][1][0]
```

```python
# Print the second element in the second nested tuples

NestedT[2][1][1]
```

<div dir =rtl><p>ما می توانیم از یک درخت برای به تصویر کشیدن این روند استفاده کنیم. هر اندکس جدید نشان دهنده یک لایه عمیق تر در درخت است:</p></div>
