###<div dir =rtl><h3><font color = blue>عملیات ست</font></h3></div>

<div dir =rtl><p>بیایید نگاهی به عملیات ست بکنیم، چون این عملیات برای تغییر ست ها به کار می رود. ست <b>A</b> را در نظر بگیرید:</p></div>


```python
# Sample set

A = set(["Thriller", "Back in Black", "AC/DC"])
A
```

<div dir =rtl><p>ما می توانیم یک جزء به ست با استفاده از متد <code>()add</code> اضافه کنیم:</p></div>

```python
# Add element to set

A.add("NSYNC")
A
```

<div dir =rtl><p>اگر ما یک جزء را دو بار اضافه کنیم اتفاقی نمی افتد. چون در ست ها مقدار تکراری نمی تواند وجود داشته باشد:</p></div>

```python
# Try to add duplicate element to the set

A.add("NSYNC")
A
```

<div dir =rtl><p>ما می توانیم یک جزء را از ست با استفاده از متد <code>remove</code> حذف نماییم:</p></div>

```python
# Remove the element from set

A.remove("NSYNC")
A
```

<div dir =rtl><p>ما می توانیم بررسی کنیم که آیا یک جزء در ست وجود دارد یا نه. این کار را با دستور <code>in</code> انجام می دهیم:</p></div>

```python
# Verify if the element is in the set

"AC/DC" in A
```

###<div dir =rtl><h3><font color = blue>عملیات منطقی ست ها</font></h3></div>

<div dir =rtl><p>به یاد داشته باشید که شما می توانید تفاوت بین ست ها را به دست آورید. همینطور فصل مشترک و اتحاد آنها را:</p></div>

<div dir =rtl><p>دو ست زیر را در نظر بگیرید:</p></div>

```python
# Sample Sets

album_set1 = set(["Thriller", 'AC/DC', 'Back in Black'])
album_set2 = set([ "AC/DC", "Back in Black", "The Dark Side of the Moon"])
```

SetsSamples.png

```python
# Print two sets

album_set1, album_set2
```

<div dir =rtl><p>از آنجا که هر دو ست حاوی <b>AC/DC</b> و <b>Back in Black</b> هستند، ما این اجزای مشترک را با فصل مشترک بین دو دایره نشان می دهیم.</p></div>

SetsLogic.png

<div dir =rtl><p>شما می توانید فصل مشترک دو ست را به صورت زیر و با استفاده از <code>&</code> به دست آورید:</p></div>

```python
# Find the intersections

intersection = album_set1 & album_set2
intersection
```

<div dir =rtl><p>شما می توانید اجزایی را که تنها در <code>album_set1</code> هستند، با استفاده از متد <code>difference</code> به دست آورید:</p></div>

```python
# Find the difference in set1 but not set2

d1 = album_set1.difference(album_set2)
```

<div dir =rtl><p>شما تنها اجزایی را که فقط در <code>album_set1</code> هستند در نظر بگیرید. تمام اجزای <code>album_set2</code> که شامل فصل مشترک هم می شود حذف می شوند.</p></div>

SetsLeft.png

<div dir =rtl><p>اجزایی که در <code>album_set2</code> هستند اما در <code>album_set1</code> نیستند را به طریق زیر به دست می آوریم:</p></div>


```python
album_set2.difference(album_set1)  
```

SetsRight.png

<div dir =rtl><p>شما همچنین می توانید فصل مشترک <code>album_list1</code> و <code>album_list2</code> را با استفاده از متد <code>intersection</code> بیابید:</p></div>

```python
# Use intersection method to find the intersection of album_list1 and album_list2

album_set1.intersection(album_set2)   
```

<div dir =rtl><p>این معادل هم پوشانی دو دایره است:</p></div>

SetsIntersect.png

<div dir =rtl><p>اتحاد نشان دهنده تمام اجزا در هر دو ست است، که با یک رنگ برای هر دو دایره نشان داده می شود:</p></div>

<img src = "https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/PY0101EN/Chapter%202/Images/SetsUnion.png" width="650" />

<div dir =rtl><p>اتحاد به این شکل به دست می آید:</p></div>

```python
# Find the union of two sets

album_set1.union(album_set2)
```

<div dir =rtl><p>شما می توانید چک کنید که آیا یک ست superset یا subset یک ست دیگر است:</p></div>

```python
# Check if superset

set(album_set1).issuperset(album_set2)   
```

```python
# Check if subset

set(album_set2).issubset(album_set1)     
```

<div dir =rtl><p>این یک مثال است که در آن <code>()issubset</code> و <code>()issuperset</code>پاسخ True می گیرند:</p></div>

```python
# Check if subset

set({"Back in Black", "AC/DC"}).issubset(album_set1) 
```

```python
# Check if superset

album_set1.issuperset({"Back in Black", "AC/DC"})   
```
