<div dir =rtl><p>شاخه دادن به ما اجازه می دهد که عبارت های متفاوتی را برای ورودی های متفاوت اجرا کنیم. این مفید است که به یک عبارت <b>if</b> به صورت یک اتاق قفل شده فکر کنیم. اگر عبارت صحیح باشد، ما می توانیم وارد اتاق شویم و برنامه ما یک سری از وظایف ازپیش تعیین شده را اجرا می نماید. ولی اگر عبارت صحیح نباشد، برنامه وظایف از پیش تعیین شده را ندیده می گیرد.</p></div>


<div dir =rtl><p>به عنوان مثال، مستطیل آبی را یک کنسرت ACDC را نمایش می دهد. اگر فرد بیش از ۱۸ سال دارد، می تواند وارد کنسرت ACDC شود. اگر آنها ۱۸ سال یا کمتر داشته باشند نمی توانند وارد کنسرت شوند. 

کد مورد نیاز برای چک کردن شرط پیشین به سادگی 
</p></div>
<code> if <i>condition statement</i> :</code>
<div dir =rtl><p>است. که شامل کلمه <code>if</code> هر عبارت شرطی و یک علامت دونقطه در آخر. کد خود را که قرار است با درست بودن شرط اجرا شود در یک خط جدید و با تو رفتگی اضافه کنید. خطوط کد بعد از علامت دو نقطه می آیند و تو رفتگی دارند فقط در صورتی اجرا می شوند که حاصل عبارت <b>if</b> برابر با <b>True</b> باشد. وظایف وقتی که کد دیگر تو رفتگی ندارد به پایان می رسند. درمورد مسله زیر، وظیفه ای که قرار است اجرا شود، <code>print(“you can enter”)</code>، فقط زمانی اجرا می شود که متغیر <code>age</code>از ۱۸ بزرگتر باشد. چون این کد دارای تورفتگی است. اما، اجرای کد <code>print(“move on”)</code> تحت تاثیر عبارت <b>if</b> قرار نمی گیرد:</p></div>

```python
# If statement example

age = 19
#age = 18

#expression that can be true or false
if age > 18:
    
    #within an indent, we have the expression that is run if the condition is true
    print("you can enter" )

#The statements after the if statement will run regardless if the condition is true or false 
print("move on")
```

<div dir =rtl><p>حالا سن برابر ۱۸ سال را در کد بالا از حالت یادداشت در آورید.</p></div>

<div dir =rtl><p>این مفید است که از نمودار زیر برای به تصویر کشیدن روند استفاده کنیم. در سمت چپ می بینیم که چه اتفاقی می افتد اگر شرط صحیح باشد. فرد وارد کنسرت ACDC می شود که بیان کننده اجرای کد درون تورفتگی است. سپس آنها به جای دیگری می روند. در قسمت راست شرایطی را می بینیم که در آن شرط صحیح نیست. فرد اجازه ورود پیدا نمی کند و به جای دیگری می رود. در این شرایط کد درون تورفتگی اجرا نمی شود اما سایر خطوط اجرا می شوند.</p></div>

<div dir =rtl><p>عبارت <code>else</code> یک بلوک از کد را اجرا می کند در صورتی که هیچ یک از شرایط پیش از این شرط صحیح نبوده باشند. بیاید از مثال کنسرت ACDC دوباره بهره بگیریم. اگر فرد ۱۷ سال داشته باشد نمی تواند به کنسرت ACDC برود اما می تواند به کنسرت Meat Loaf برود. 

دستور عبارت <code>else</code> شبیه دستور عبارت <code>if</code> است، یعنی  <code>: else</code> توجه کنید که شرطی در مورد عبارت<code>else</code> وجود ندارد، بر خلاف عبارت <code>if</code>.

مقدار متغیر <code>age</code> را تغییر دهید تا ببینید چه اتفاقی می افتد: </p></div>

```python
# Else statement example

age = 18
# age = 19

if age > 18:
    print("you can enter" )
else:
    print("go see Meat Loaf" )
    
print("move on")
```

<div dir =rtl><p>این روند در شکل زیر نمایش داده شده است، هر یک از احتمالات در یک طرف شکل نشان داده شده اند. در قسمت چپ حالتی است که سن برابر با ۱۷ است. در نتیجه فرد به کنسرت Meat Loaf می رود. در قسمت راست و وقتی سن بیش از ۱۸ است به فرد اجازه دسترسی به کنسرت داده می شود. </p></div>

<div dir =rtl><p>عبارت <code>elif</code> که مخفف شده  else if است، به ما اجازه می دهد شروط بیشتری را چک کنیم اگر شروط پیش از آن درست نبوده باشند. اگر شرط  عبارت <code>elif</code> صحیح باشد، کد های جایگزین اجرا می شوند. مثال کنسرت را در نظر بگیرید. فردی با سن ۱۸ سال وارد می شود. از آنجایی که بیش از ۱۸ سال ندارد نمی تواند وارد کنسرت شود اما چون ۱۸ سال دارد می تواند به کنسرت Pink Floyd برود. بعد از دیدن Pink Floyd به جای دیگری می رود. دستور عبارت <code>elif</code> کاملا شبیه عبارت <code>if</code> است و فقط به جای <code>if</code> از <code>elif</code> استفاده می شود.</p></div>


```python
# Elif statment example

age = 18

if age > 18:
    print("you can enter" )
elif age == 18:
    print("go see Pink Floyd")
else:
    print("go see Meat Loaf" )
    
print("move on")
```

```python
# Condition statement example

album_year = 1990

if(album_year < 1980) or (album_year > 1989):
    print ("Album was not made in the 1980's")
else:
    print("The Album was made in the 1980's ")
```

The <code>not</code> statement checks if the statement is false:
<div dir =rtl><p>عبارت <code>not</code> چک می کند که آیا عبارت صحیح است.</p></div>

```python
# Condition statement example

album_year = '1984'

if not (album_year == '1984'):
    print ("Album year is not 1984")
else:
    print ("Album year is 1984")
album_year
```
