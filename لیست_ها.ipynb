<div dir =rtl><p>از آنجا که لیست ها جهش پذیر هستند، می توانیم آنها را تغییر دهیم. به عنوان مثال می توانیم اولین جزء را به طریق زیر تغییر دهیم:</p></div>

```python
# Change the element based on the index

A = ["disco", 10, 1.2]
print('Before change:', A)
A[0] = 'hard rock'
print('After change:', A)
```

<div dir =rtl><p>می توانیم  یک جزء از لیست را با دستور <code>del</code> حذف نماییم:</p></div>

```python
# Delete the element based on the index

print('Before change:', A)
del(A[0])
print('After change:', A)
```

<div dir =rtl><p>ما می توانیم یک رشته حروف را با استفاده از <code>split</code> به لیست تبدیل کنیم. به عنوان مثال این دستور هر گروه از حروف را که با یک فاصله از هم حدا شده اند به یک جزء از لیست تبدیل می کند:</p></div>

```python
# Split the string, default is by space

'hard rock'.split()
```

<div dir =rtl><p>ما می توانیم از دستور <code>split</code>استفاده کنیم تا رشته حروف را در حرفی خاص از هم جدا کنیم. برای این کار ، حرفی را که می خواهیم با رسیدن به آن تکه شدن اتفاق بیفتد را به عنوان یک نشانوند به دستور ذکر شده می دهیم (در این مورد کاما). نتیجه یک لیست است و هر جزء  تشکیل شده است از حروفی که با کاما از هم جدا شده اند:</p></div>


```python
# Split the string by comma

'A,B,C,D'.split(',')
```

###<div dir =rtl><h3><font color = blue>کپی کردن و کلون کردن لیست</font></h3></div>

<div dir =rtl><p>وقتی ما یک متغیر <b>B</b> را برابر با <b>A</b> قرار می دهیم، هر دو در حافظه به یک لیست اشاره دارند:</p></div>

```python
# Copy (copy by reference) the list A

A = ["hard rock", 10, 1.2]
B = A
print('A:', A)
print('B:', B)
```

ListsRef.png

<div dir =rtl><p>در ابتدا مقدار اولین جزء در<b>B</b> برابر با hard rock است. اگر ما اولین جزء در <b>A</b> را به <b>banana</b> تغییر دهیم، یک اثر جانبی غیر منتظره می گیریم. از آنجا که <b>A</b> و <b>B</b> به یک لیست واحد اشاره دارند> اگر لیست <b>A</b> تغییر کند <b>B</b> هم تغییر می کند. اگر ما جزء اول <b>B</b> را چک کنیم به جای hard rock کلمه banana را می گیریم:</p></div>


```python
# Examine the copy by reference

print('B[0]:', B[0])
A[0] = "banana"
print('B[0]:', B[0])
```

<div dir =rtl><p>شما می توانید لیست <b>A</b> را با استفاده از دستور زیر کلون کنید:</p></div>

```python
# Clone (clone by value) the list A

B = A[:]
B
```

<div dir =rtl><p>متغیر <b>‌B</b> به یک کپی جدید از لیست اصلی ارجاع می دهد. این در شکل زیر نمایس داده شده است:</p></div>

<div dir =rtl><p>حالا اگر <b>A</b> را تغییر دهیم، <b>B</b> تغییر نمی کند:</p></div>

```python
print('B[0]:', B[0])
A[0] = "hard rock"
print('B[0]:', B[0])
A
```
