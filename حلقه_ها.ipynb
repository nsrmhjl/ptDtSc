###<div dir =rtl><h3><font color = blue>حلقه <code>for</code> چیست؟</font></h3></div>

<div dir =rtl><p>حلقه <code>for</code> شما را قادر می کند که یک بلوک از کد را چندین بار اجرا کنید. به عنوان مثال، شما از آن استفاده میکنید تا تک تک اجزای یک لیست را چاپ کنید.</p></div>
<div dir =rtl><p>بیایید از یک حلقه <code>for</code> استفاده کنیم تا سالهای موجود در لیست <code>dates</code> را چاپ کنیم:</p></div>

<div dir =rtl><p>این کار به طریق زیر انجام می شود:</p></div>

```python
# For loop example

dates = [1982,1980,1973]
N = len(dates)

for i in range(N):
    print(dates[i])     
```

<div dir =rtl><p>کد درون تورفتگی <code>N</code> بار اجرا می شود. هر بار مقدار <code>i</code> بعد از هر بار اجرا یکی زیاد می شود. عبارتی که اجرا می شود <code>print</code> است برای هر جزء از لیست که دارای اندکس <code>i</code> است. همانطور که در اینجا مشاهده می شود:</p></div>

LoopsForRange.gif

<div dir =rtl><p>در این مثال ما یک توالی از اعداد بین صفر تا ۷ را چاپ کنیم:</p></div>

```python
# Example of for loop

for i in range(0, 8):
    print(i)
```

<div dir =rtl><p>در پایتون ما می توانیم به طور مستقیم به اجزای لیست به طریق زیر دسترسی پیدا کنیم:</p></div>

```python
# Exmaple of for loop, loop through list

for year in dates:  
    print(year)   
```

<div dir =rtl><p>برای هر تکرار، مقدار متغیر <code>years</code> مانند مقدار <code>dates[i]</code> در مثال اول عمل می کند:</p></div>

<div dir =rtl><p>ما می توانیم اجزای لیست را تغییر دهیم:</p></div>

```python
# Use for loop to change the elements in list

squares = ['red', 'yellow', 'green', 'purple', 'blue']

for i in range(0, 5):
    print("Before square ", i, 'is',  squares[i])
    squares[i] = 'white'
    print("After square ", i, 'is',  squares[i])
```

<div dir =rtl><p>ما می توانیم به اندکس و اجزای لیست به طریق زیر دسترسی یابیم:</p></div>

```python
# Loop through the list and iterate on both index and element value

squares=['red', 'yellow', 'green', 'purple', 'blue']

for i, square in enumerate(squares):
    print(i, square)
```

###<div dir =rtl><h3><font color = blue>حلقه <code>while</code> چیست؟</font></h3></div>

<div dir =rtl><p>همانطور که دیدید، حلقه <code>for</code> برای یک جریان کنترل شده از تکرار استفاده می شود. اما، اگر ما ندانیم که کی می خواهیم حلقه را متوقف کنیم چه؟ اگر ما می خواستیم یک بلوک از کد را اجرا کنیم تا شرطی صحیح باشد چه؟ حلقه <code>while</code> ابزاری برای تکرار بر اساس یک شرط است. بلوک کد مکرر اجرا می شود تا وقتی که شرط منطقی مقدار دوتایی False را تولید کند.</p></div>

<div dir =rtl><p>فرض کنید می خوهیم درون لیست <code>dates</code> تکرار کنیم تا به سال ۱۹۷۳ برسیم و سپس تعداد تکرار را چاپ کنیم. این امر با بلوک کد زیر انجام پذیر است:</p></div>

```python
# While Loop Example

dates = [1982, 1980, 1973, 2000]

i = 0
year = 0

while(year != 1973):
    year = dates[i]
    i = i + 1
    print(year)

print("It took ", i ,"repetitions to get out of loop.")
```

<div dir =rtl><p>یک حلقه while اجرای کد را تکرار می کند تا زمانی که شرط نشانوند صحیح نباشد، همانطور که در تصویر زیر نشان داده شده:</p></div>
