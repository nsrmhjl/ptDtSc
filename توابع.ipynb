<div dir =rtl><p>تا اینجا، ما متغیرهایی در درون توابع به وجود می آوردیم، اما متغیر های بیرون توابع را تا حالا بحث نکرده ایم. این متغیر ها جهانی نامیده می شوند.</p></div>

```python
# Example of global variable

artist = "Michael Jackson"
def printer1(artist):
    internal_var = artist
    print(artist, "is an artist")
    
printer1(artist)
```

<div dir =rtl><p>اگر ما <code>internal_var</code> را چاپ نماییم یک پیغام خطا می گیریم.</p></div>


<b>We got a Name Error:  <code>name 'internal_var' is not defined</code>. Why?</b>  

It's because all the variables we create in the function is a <b>local variable</b>, meaning that the variable assignment does not persist outside the function.  

But there is a way to create <b>global variables</b> from within a function as follows:

<div dir =rtl><p>ما یک پیغام خطای اسم می گیریم: <code>name 'internal_var' is not defined</code> به چه دلیل؟</p></div>

<div dir =rtl><p>اما برای تعریف متغیر های جهانی از درون تابع راهی وجود دارد:</p></div>





```python
artist = "Michael Jackson"

def printer(artist):
    global internal_var 
    internal_var= "Whitney Houston"
    print(artist,"is an artist")

printer(artist) 
printer(internal_var)
```

##<div dir =rtl><h2><font color = blue>محدوده متغیر</font></h2></div>



<div dir =rtl><p>محدوده متغیر قسمتی از برنامه است که آن متغیر در آن قسمت در دسترس قرار دارد. متغیر هایی که بیرون از تعریف توابع تعریف می شوند، مثل <code>myFavouriteBand</code> که در کد زیر نشان داده شده، در همه جای برنامه در دسترس قرار دارند. به همین دلیل گفته می شود که این متغیر ها محدوده جهانی دارند. و به عنوان متغیر جهانی شناخته می شوند.</p></div>
<div dir =rtl><p><code>myFavouriteBand</code> یک متغیر جهانی است پس ان متغیر از داخل تابع <code>getBandRating</code> قابل دسترسی است تا ما از آن برای تعیین امتیاز گروه موسیقی استفاده نماییم. ما همچنین می توانیم از آن در بیرون از تابع نیز استفاده نماییم و آن را به تابع print رد کنیم تا نمایش دهد: </p></div>


```python
# Example of global variable

myFavouriteBand = "AC/DC"

def getBandRating(bandname):
    if bandname == myFavouriteBand:
        return 10.0
    else:
        return 0.0

print("AC/DC's rating is:", getBandRating("AC/DC"))
print("Deep Purple's rating is:",getBandRating("Deep Purple"))
print("My favourite band is:", myFavouriteBand)
```

<div dir =rtl><p>به این ویرایش تغییر یافته از کد  بالایی نگاهی بیندازید. اکنون متغیر <code>myFavouriteBand</code> در درون تابع <code>getBandRating</code> تعریف شده است. متغیری که در درون یک تابع تعریف شده است، گفته می شود که متغیر محلی آن تابع است. این به آن معنی است که این متغیر تنها در درون تابعی که در آن تعریف شده است قابل دسترسی است. تابع <code>getBandRating</code> همچنان کار می کند. چون متغیر <code>myFavouriteBand</code> همچنان در درون تابع قابل دسترسی است. اما ما دیگر نمی توانیم <code>myFavouriteBand</code> را بیرون از تابع چاپ نماییم. چون یک متغیر محلی تابع <code>getBandRating</code>  بوده و تنها از درون آن قابل دسترسی است:</p></div>


```python
# Example of local variable

def getBandRating(bandname):
    myFavouriteBand = "AC/DC"
    if bandname == myFavouriteBand:
        return 10.0
    else:
        return 0.0

print("AC/DC's rating is: ", getBandRating("AC/DC"))
print("Deep Purple's rating is: ", getBandRating("Deep Purple"))
print("My favourite band is", myFavouriteBand)
```

<div dir =rtl><p>در پایان نگاهی به این مثال بیاندازید. حالا ما در نوع تعریف <code>myFavouriteBand</code> داریم. اولی یک محدوده جهانی دارد و دومی یک محدوده محلی در تابع <code>getBandRating</code>. در درون تابع متغیر محلی پیش دست است. <b>Deep Purple</b> اگر به تابع <code>getBandRating</code> رد شود امتیاز ۱۰ خواهد گرفت. اما بیرون از تابع <code>getBandRating</code>، متغیر محلی <code>getBandRating</code> تعریف نشده است. پس متغیر <code>myFavouriteBand</code> که ما آن را چاپ می کنیم یک متغیر جهانی است که مقدار آن <b>AC/DC</b> است.</p></div>


```python
# Example of global variable and local variable with the same name

myFavouriteBand = "AC/DC"

def getBandRating(bandname):
    myFavouriteBand = "Deep Purple"
    if bandname == myFavouriteBand:
        return 10.0
    else:
        return 0.0

print("AC/DC's rating is:",getBandRating("AC/DC"))
print("Deep Purple's rating is: ",getBandRating("Deep Purple"))
print("My favourite band is:",myFavouriteBand)
```
