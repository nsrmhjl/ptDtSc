##<div dir =rtl><h2><font color = blue>توالی فرار </font></h2></div>

<div dir =rtl><p>Back slashe نماینده شروع یک توالی فرار یا escape sequence است. توالی های نماینده رشته حروفی هستند که ورود آنها مشکل است. به عنوان مثال back slash "n" نشان دهنده شروع یک خط جدید است. در هنگام دریافت این رشته حروف، خروجی به یک خط جدید می رود: </p></div>

```python
# New line escape sequence

print(" Michael Jackson \n is the best" )
```

<div dir =rtl><p>به صورت مشابه back slash  "t" نماینده tab است: </p></div>

```python
# Tab escape sequence

print(" Michael Jackson \t is the best" )
```

<div dir =rtl><p>اگر می خواهید یک back slash در رشته حروف خود قرار دهید، از علامت دوگانه back slash استفاده کنید:</p></div>

```python
# Include back slash in string

print(" Michael Jackson \\ is the best" )
```

<div dir =rtl><p>برای آنکه backslash به نمایش درآید ما همچنین می توانیم از یک "r" در ابتدای رشته حروف استفاده نماییم:</p></div>

```python
# r will tell python that string will be display as raw string

print(r"Michael Jackson \n is the best" )
```

##<div dir =rtl><h2><font color = blue>عملیات رشته حروف</font></h2></div>


<div dir =rtl><p>تعداد زیادی عملیات رشته حروف در پایتون وجود دارد که می توان از آنها برای دستکاری داده استفاده نمود. ما قصد داریم تا از چند عملیات پایه ای بر روی داده استفاده کنیم:</p></div>

<div dir =rtl><p>بیایید متد<code>upper</code>را امتحان کنیم؛ این متد حروف کوچک را به حروف بزرگ تبدیل می کند: </p></div>


```python
# Convert all the characters in string to upper case

A = "Thriller is the sixth studio album"
print("before upper:", A)
B = A.upper()
print("After upper:", B)
```

<div dir =rtl><p>متد <code>replace</code> قسمتی از یک رشته حروف را که اصطلاحا substring خوانده می شود با یک رشته حروف دیگر جایگزین می کند. ما قسمتی از رشته حروف را که می خواهیم تغییر کند وارد می کنیم. قسمت دوم رشته حروفی است که می خواهیم قسمت مورد نظر را با آن تعویض نماییم، و نتیجه رشته حروف جدیدی است که قسمتی از آن به گونه ای که ما خواسته ایم تغییر نموده است:</p></div>


```python
# Replace the old substring with the new target substring is the segment has been found in the string

A = "Michael Jackson is the best"
B = A.replace('Michael', 'Janet')
B
```

<div dir =rtl><p>متد<code>find</code>یک substring را پیدا می کند. ورودی قسمتی است که شما مایلید پیدا کنید و خروجی اولین اندکس قسمت یافته شده است. ما می توانیم <code>jack</code>یا<code>el</code>را پیدا کنیم.</p></div>


StringsFind.png

```python
# Find the substring in the string. Only the index of the first elment of substring in string will be the output

Name = "Michael Jackson"
Name.find('el')
```

```python
# Find the substring in the string.

Name.find('Jack')
```

<div dir =rtl><p>اگر قسمتی که به دنبال آن می گردید در رشته حروف مورد نظر وجود نداشته باشد، خروجی منفی یک (۱-) خواهد بود. به عنوان مثال 'Jasdfasdasdf' یافته نخواهد شد:</p></div>


```python
# If cannot find the substring in the string

Name.find('Jasdfasdasdf')
```
