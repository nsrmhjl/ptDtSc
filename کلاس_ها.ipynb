<div dir =rtl><p><i>شکل ۴: روش ساختن یک شیء</i></p></div>


<div dir =rtl><p>مرحله بعد یک متد خاص است با نام بنا کننده یا constructor که با <code>&#95;&#95;init&#95;&#95;</code> نشان داده می شود و برای شروع کردن شیء به کار می رود. ورودی ها ویژگی های داده یا data attribute هستند. کلمه <code>self</code> تمام ویژگی ها را در خود جای می دهد. به عنوان مثال <code>self.color</code> به ما مقدار ویژگی رنگ را می دهد و <code>self.radius</code> به ما مقدار ویژگی شعاع را می دهد. ما همچنین متد <code>()add_radius</code> را داریم به همراه متغیر <code>r</code>. این متد مقدار متغیر <code>r</code> را به ویژگی شعاع اضافه می کند. برای دسترسی به شعاع ما از دستور <code>self.radius</code> استفاده می کنیم. دستور نشان دار شده در شکل ۵ خلاصه شده است:</p></div>

ClassesCircle.png

<div dir =rtl><p><i>شکل شماره ۵: دستور نشاندار شده شیء دایره.</i></p></div>


<div dir =rtl><p>شیء اصلی در زیر نمایش دهده شده است. ما متد <code>drawCircle</code> را هم اضافه کرده ایم تا عکس دایره را رسم نماییم. ما شعاع پیش فرض را برابر ۳ و رنگ پیش فرض را برابر آبی قرار داده ایم.</p></div>


```python
# Create a class Circle

class Circle(object):
    
    # Constructor
    def __init__(self, radius=3, color='blue'):
        self.radius = radius
        self.color = color 
    
    # Method
    def add_radius(self, r):
        self.radius = self.radius + r
        return(self.radius)
    
    # Method
    def drawCircle(self):
        plt.gca().add_patch(plt.Circle((0, 0), radius=self.radius, fc=self.color))
        plt.axis('scaled')
        plt.show()  
```

##<div dir =rtl><h2><font color = blue>کلاس مستطیل</font></h2></div>


<div dir =rtl><p>بیایید یک کلاس مستطیل با ویژگی های ارتفاع، عرض و رنگ بسازیم. ما فقط متدی اضافه می کنیم که با آن بتوانیم مستطیل را رسم نماییم:</p></div>


```python
# Create a new Rectangle class for creating a rectangle object

class Rectangle(object):
    
    # Constructor
    def __init__(self, width=2, height=3, color='r'):
        self.height = height 
        self.width = width
        self.color = color
    
    # Method
    def drawRectangle(self):
        plt.gca().add_patch(plt.Rectangle((0, 0), self.width, self.height ,fc=self.color))
        plt.axis('scaled')
        plt.show()
```

<div dir =rtl><p>بیایید شیء <code>SkinnyBlueRectangle</code> را از نوع مستطیل ببه وجود آوریم. عرض آن ۲ و ارتفاع آن ۳ و رنگ آن آبی خواهد بود:</p></div>


```python
# Create a new object rectangle

SkinnyBlueRectangle = Rectangle(2, 10, 'blue')
```
