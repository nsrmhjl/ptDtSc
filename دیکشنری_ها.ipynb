###<div dir =rtl><h3><font color = blue>دیکشنری ها چه هستند؟</font></h3></div>

<div dir =rtl><p>یک دیکشنری از کلید ها و مقادیر تشکیل شده است. مقایسه دیکشنری با لیست مفید است. به جای اندکس های عددی در لیست، دیکشنری ها کلید دارند. این کلید ها برای دسترسی به مقادیر دیکشنری استفاده می شوند.</p></div>

DictsList.png

<div dir =rtl><p>یک مثال از دیکشنری <code>Dict</code>:</p></div>

```python
# Create the dictionary

Dict = {"key1": 1, "key2": "2", "key3": [3, 3, 3], "key4": (4, 4, 4), ('key5'): 5, (0, 1): 6}
Dict
```

<div dir =rtl><p>کلید ها می توانند رشته حروف باشند:</p></div>

```python
# Access to the value by the key

Dict["key1"]
```

<div dir =rtl><p>کلید ها همینطور می توانند هر شیء غیر قابل جهش دیگر مانند تاپل باشند:</p></div>

```python
# Access to the value by the key

Dict[(0, 1)]
```

<div dir =rtl><p>هر کلید از مقدار خود با یک نشان دو نقطه جدا می شود ("<code>:</code>"). هر ورودی با کاما از ورودی دیگر جدا می شود و کل دیکشنری با آکولاد  محصور می شود مانند این "<code>{}</code>".</p></div>

```python
# Create a sample dictionary

release_year_dict = {"Thriller": "1982", "Back in Black": "1980", \
                    "The Dark Side of the Moon": "1973", "The Bodyguard": "1992", \
                    "Bat Out of Hell": "1977", "Their Greatest Hits (1971-1975)": "1976", \
                    "Saturday Night Fever": "1977", "Rumours": "1977"}
release_year_dict
```

<div dir =rtl><p>به طور خلاصه، مانند یک لیست، یک دیکشنری حاوی یک توالی از اجزا است. هر جزء با یک کلید و مقدار متناظر آن نمایندگی می شود. دیکشنری ها  با آکولاد درست می شوند که درون آنها کلید ها و مقادیری قرار دارند که با نشان دو نقطه از هم جدا می شوند. برای هر کلید تنها یک مقدار می تواند وجود داشته باشد، اما اما چند کلید می توانند مقداری یکسان داشته باشند. کلید ها تنها می توانند اعداد، رشته حروف و تاپل ها باشند اما مقادیر می توانند هر نوعی از انواع داده باشند.</p></div>


<div dir =rtl><p>مفید است که دیکشنری را به صورت یک جدول تصویر نماییمو که در شکل زیر می بینید. اولین ستون نماینده کلید ها و دومین ستون نماینده مقادیر است.</p></div>
